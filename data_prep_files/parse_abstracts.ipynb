```python
# 2024

import fitz  # PyMuPDF  
import re  
import os  
  
# Path to your PDF file  
pdf_path = "HCSF 2024_ Book of Abstracts.pdf"  
  
# Open the PDF file  
doc = fitz.open(pdf_path)  
  
# Extract all text from the document  
full_text = ""  
for page in doc:  
    full_text += page.get_text()  
  
# Close the PDF document  
doc.close()  
  
# Define regex pattern to find abstract codes (e.g., "1.A", "216.J")  
code_pattern = r"\b\d+\.[A-Z]\b"  
  
# Find all abstract codes and their positions  
codes = [(m.group(), m.start()) for m in re.finditer(code_pattern, full_text)]  
  
# Sort codes by their position in text  
codes.sort(key=lambda x: x[1])  
  
# Create output directory named "2024_abstracts"  
output_dir = "2024_abstracts"  
os.makedirs(output_dir, exist_ok=True)  
  
# Loop through the abstract codes and extract text between consecutive codes  
for idx, (code, start_pos) in enumerate(codes):  
    end_pos = codes[idx + 1][1] if idx + 1 < len(codes) else len(full_text)  
    abstract_text = full_text[start_pos:end_pos].strip()  
  
    # Safe filename (replace '.' with '_')  
    safe_code = code.replace(".", "_")  
  
    # File path to save abstract  
    file_path = os.path.join(output_dir, f"{safe_code}.txt")  
  
    # Save abstract text to file  
    with open(file_path, "w", encoding="utf-8") as file:  
        file.write(abstract_text)  
  
    print(f"Saved abstract {code} to '{file_path}'")  
  
print("All abstracts extracted successfully.")  
```

```python
# 2018

import fitz  # PyMuPDF  
import re  
import os  
  
# Path to your PDF file  
pdf_path = "HCSF 2018_ Book of Abstracts.pdf"  
  
# Open the PDF file  
doc = fitz.open(pdf_path)  
  
# Extract all text from the document  
full_text = ""  
for page in doc:  
    full_text += page.get_text()  
  
doc.close()  
  
# Define regex pattern to find abstract numbers (e.g., "1", "2", "3", etc.) at the start of abstracts  
# The abstracts seem to be numbered clearly as "Abstract 1", "Abstract 2", etc.  
# Therefore, our pattern will look explicitly for "Abstract 1", "Abstract 2", etc.  
code_pattern = r"\bAbstract\s+(\d+)\b"  
  
# Find all abstract numbers and their positions  
codes = [(m.group(1), m.start()) for m in re.finditer(code_pattern, full_text)]  
  
# Sort codes by position  
codes.sort(key=lambda x: x[1])  
  
# Create output directory named "2018_abstracts"  
output_dir = "2018_abstracts"  
os.makedirs(output_dir, exist_ok=True)  
  
# Loop through abstract numbers and extract text between consecutive abstracts  
for idx, (code, start_pos) in enumerate(codes):  
    end_pos = codes[idx + 1][1] if idx + 1 < len(codes) else len(full_text)  
    abstract_text = full_text[start_pos:end_pos].strip()  
  
    # Filename based on abstract number  
    safe_code = code  
  
    # File path to save abstract  
    file_path = os.path.join(output_dir, f"{safe_code}.txt")  
  
    # Save abstract text to file  
    with open(file_path, "w", encoding="utf-8") as file:  
        file.write(abstract_text)  
  
    print(f"Saved abstract {code} to '{file_path}'")  
  
print("All 2018 abstracts extracted successfully.")  
```

```python
import fitz  # PyMuPDF  
import re  
import os  
  
# Path to your PDF file  
pdf_path = "HCSF 2017_ Book of Abstracts.pdf"  
  
# Open the PDF file  
doc = fitz.open(pdf_path)  
  
# Extract all text from the document  
full_text = ""  
for page in doc:  
    full_text += page.get_text()  
  
doc.close()  
  
# Define regex pattern to find abstract numbers and titles  
# Pattern explanation:  
#   - ^\s*(\d{3}) matches three-digit abstract numbers at the beginning of a line (allowing for spaces)  
#   - \s+(.+) matches at least one space and captures the rest of the line (the abstract title)  
pattern = re.compile(r"^\s*(\d{3})\s+(.+)", re.MULTILINE)  
  
# Find all matches and positions  
matches = [(m.group(1), m.start()) for m in re.finditer(pattern, full_text)]  
  
# Sort matches by position  
matches.sort(key=lambda x: x[1])  
  
# Create output directory named "2017_abstracts"  
output_dir = "2017_abstracts"  
os.makedirs(output_dir, exist_ok=True)  
  
# Loop through abstract matches and extract text between consecutive abstracts  
for idx, (abstract_number, start_pos) in enumerate(matches):  
    end_pos = matches[idx + 1][1] if idx + 1 < len(matches) else len(full_text)  
    abstract_text = full_text[start_pos:end_pos].strip()  
  
    # File path to save abstract  
    file_path = os.path.join(output_dir, f"{abstract_number}.txt")  
  
    # Save abstract text to file  
    with open(file_path, "w", encoding="utf-8") as file:  
        file.write(abstract_text)  
  
    print(f"Saved abstract {abstract_number} to '{file_path}'")  
  
print("All 2017 abstracts extracted successfully.")  
```

```python
# 2016

import fitz  # PyMuPDF  
import re  
import os  
  
# Path to your PDF file  
pdf_path = "HCSF 2016_ Book of Abstracts.pdf"  
  
# Open the PDF file  
doc = fitz.open(pdf_path)  
  
# Extract all text from the document  
full_text = ""  
for page in doc:  
    full_text += page.get_text()  
  
doc.close()  
  
# Improved regex pattern explanation:  
# ^\s*(\d\.\d{2})\s+[A-Z] matches abstract numbers (e.g., 1.01) at the start of lines,  
# followed by at least one whitespace and then an uppercase letter,  
# indicating the beginning of the abstract title.  
code_pattern = re.compile(r"^\s*(\d\.\d{2})\s+[A-Z]", re.MULTILINE)  
  
# Find all abstract numbers and positions using improved regex  
matches = [(m.group(1), m.start()) for m in code_pattern.finditer(full_text)]  
  
# Sort matches by position  
matches.sort(key=lambda x: x[1])  
  
# Create output directory named "2016_abstracts"  
output_dir = "2016_abstracts"  
os.makedirs(output_dir, exist_ok=True)  
  
# Loop through abstract matches and extract text between consecutive abstracts  
for idx, (abstract_number, start_pos) in enumerate(matches):  
    end_pos = matches[idx + 1][1] if idx + 1 < len(matches) else len(full_text)  
    abstract_text = full_text[start_pos:end_pos].strip()  
  
    # File path to save abstract  
    file_path = os.path.join(output_dir, f"{abstract_number}.txt")  
  
    # Save abstract text to file  
    with open(file_path, "w", encoding="utf-8") as file:  
        file.write(abstract_text)  
  
    print(f"Saved abstract {abstract_number} to '{file_path}'")  
  
print("All 2016 abstracts extracted successfully.")
```
